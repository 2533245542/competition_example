```python
% matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df663 = pd.read_csv('663dataMultivariateLSTM.csv')
df663['date'] = pd.to_datetime( df663[['year', 'month', 'day']].apply(lambda x: '-'.join([str(i) for i in x]), axis=1)).rename('date')
df663 = df663.set_index(['date'], drop=True)
df663[lambda df: df['dayofweek'] != 5][lambda df: df['dayofweek'] != 6][['n_walkins']][:].plot(style='-', title= '663')

```
